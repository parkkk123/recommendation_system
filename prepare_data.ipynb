# Cleaning Data

## Check data quality from rating

```python
#Ratings are made on a 5-star scale, with half-star increments (0.5 stars - 5.0 stars). 
df[(df.rating < 0.5) | (df.rating  > 5)]
```

```python
df =  df[~(df.rating < 0.5) | (df.rating  > 5)]
df
```

## Drop null users that impact to recommend

```python
# Drop Null User
df = df[~pd.isnull(df.userId)]
df
```
